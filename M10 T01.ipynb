#### Scraping de Bolsa Madrid fent servir Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")
```

```python
url = 'https://www.bolsasymercados.es/bme-exchange/es/Mercados-y-Cotizaciones/Acciones/Mercado-Continuo/Precios/ibex-35-ES0SI0000005'

driver.get(url)
```

```python
rows = []
for table in driver.find_elements(by= By.XPATH, value = '//*[contains(@class,"shares-table")]//tr'):
        data = [item.text for item in table.find_elements(by= By.XPATH, value= ".//*[self::td or self::th]")]
        # afegim les dades per files
        rows.append(data)
        
# la primera filera és la que pertany als noms de les columnes.
stock_indices = pd.DataFrame(rows[1:], columns = rows[0])
stock_indices
```

#### Fem el webscraping de la wikipedia fent servir BeautifulSoup

```python
# Request de la pagina web
import requests as r

wiki_page_request = r.get("https://en.wikipedia.org/wiki/Comparison_of_reference_management_software")
wiki_page_text = wiki_page_request.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(wiki_page_text, 'html.parser')
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.a
```

```python
# Aquí busquem totes les taules
soup.find_all('table')

# I mostrem el numero de taules
len(soup.find_all('table'))
```

```python
# Extareiem la primera taula
table_soup = soup.find('table')
```

```python
# Obtenim els encapçalaments buscant tots el 'th'
headers = [header.text.strip() for header in table_soup.find_all('th')]
```

```python
headers
```

#### Ens ha unit tots els encapçalaments de columnes amb els encapçalaments de les fileres

```python
# Separem els encapçalaments de columnes del de les fileres
Column_names = headers[0:9]
first_row = headers[9:]
```

```python
# Ara treurem tots els valors de les cel·les
rows = []

# Trobem tots els 'tr'
data_rows = table_soup.find_all('tr')

# Extraiem totes les cel·les buscant els 'td'
for row in data_rows:
    value = row.find_all('td')
    beautified_value = [ele.text.strip() for ele in value]
    # Eliminen si estan buits
    if len(beautified_value) == 0:
        continue
    rows.append(beautified_value)
```

```python
# I finalment monten un DataFrame amb tots els elements que hem extret

import pandas as pd

df = pd.DataFrame(rows)
df.insert(0, 'val',first_row)
df.columns = Column_names
df
```
